##### This SQL query will join all five tables and sort them by Last and First Name

```python
sql = '''SELECT name.LastName, name.FirstName, track.Name, alb.Title
         FROM Customer as name
         LEFT JOIN Invoice as invo
         ON name.CustomerId = invo.CustomerId
         LEFT JOIN InvoiceLine as involine
         ON invo.InvoiceId = involine.InvoiceId
         LEFT JOIN Track as track
         ON involine.TrackId = track.TrackId
         LEFT JOIN Album as alb
         ON track.AlbumId = alb.AlbumId
         ORDER BY name.LastName, name.FirstName'''

con = sqlite3.connect("Chinook_Sqlite.sqlite")
db = pd.read_sql_query(sql , con=con).head(10)
con.close()
db
```
